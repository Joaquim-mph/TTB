```python
import numpy as np
from scipy.optimize import curve_fit
from uncertainties import ufloat
import matplotlib.pyplot as plt
# Define the Gaussian function
def gaussian(x, mu, sigma, amplitude):
    return amplitude * np.exp(-(x - mu)**2 / (2 * sigma**2))

# Generate some data with noise
x_data = np.linspace(-5, 5, num=50)
y_data = gaussian(x_data, 0, 1, 1) + np.random.normal(0, 0.1, len(x_data))

# Fit the data to the Gaussian function
popt, pcov = curve_fit(gaussian, x_data, y_data)

# Get the fitted y values
y_fit = gaussian(x_data, *popt)

# Calculate the residuals
residuals = y_data - y_fit

# Calculate the reduced chi-squared
chi_squared = np.sum((residuals / y_data) ** 2) / (len(x_data) - len(popt))
reduced_chi_squared = chi_squared / (len(x_data) - len(popt))

print("Reduced chi-squared:", reduced_chi_squared)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots(2, 1, figsize=(10, 8))

ax[0].plot(x, y1)
ax[0].set_title("Sin Plot")
ax[0].set_xlabel("X")
ax[0].set_ylabel("Y")

ax[1].plot(x, y2)
ax[1].set_title("Cos Plot")
ax[1].set_xlabel("X")
ax[1].set_ylabel("Y")

plt.tight_layout()
plt.show()
```
